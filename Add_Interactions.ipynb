```python
import pandas as pd
```

```python
df = pd.read_csv('hackathon_kpis_anonymised/kpis_cleaned_notonehot_median.csv')
df.drop(columns='Unnamed: 0', inplace=True)
```

```python
list(df.columns)
```

```python
interactions = {
    'voice': ('voice_drop_rate', 'num_voice_attempts'),
    'data': ('data_drop_rate', 'num_data_attempts'),
    'voice_setup': ('voice_setup_failure_rate', 'num_voice_attempts'),
}
```

```python
for key, (up, down) in interactions.items():
    df[key] = df[up] - df[down]
```
