```python
%load_ext autoreload
%autoreload 2
from plot import *
import pandas as pd
```

```python
df = pd.read_csv("hackathon_kpis_anonymised/ml_models_dataset_with_zone_v2.csv")
```

```python
plot_cols = [
    'ho_failure_rate', 'num_voice_attempts',
       'voice_drop_rate', 'num_data_attempts', 'voice_setup_failure_rate',
       'voice_tot_failure_rate', 'avail_period_duration', 'bandwidth',
       'throughput_rate', 'data_setup_failure_rate', 'data_drop_rate'
]
```

```python
def plot_results(results_path='preds/deepant_00_11R.csv', cell_name='00_11R', threshold=0.01):
    preds = pd.read_csv(results_path) if '.csv' in results_path else pd.read_pickle(results_path)
    
    if 'Unnamed: 0' in preds.columns:
        print('Set "Unnamed: 0" as index')
        preds.set_index('Unnamed: 0', inplace=True)
    
    if 'cell_name' in preds.columns:
        print('Filter preds by cell name')
        cell_preds = preds.loc[preds['cell_name'] == cell_name]
    else:
        print('Not filtering, assuming preds only contain', cell_name)
        cell_preds = preds

    cell_df = df.loc[df['cell_name']==cell_name]

    plot_anomalies(cell_df, cell_preds, plot_cols, title=cell_name, threshold=threshold)

#plot_results()
```

```python
plot_results('preds/std_baseline_anomalies.csv', '00_11R')
```
