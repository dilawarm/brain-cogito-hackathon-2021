```
def calculate_clusters(data):
    labels = list(data.columns)

    # Perform clustering, you can choose the method
    # in this case, we use 'ward'
    Z = linkage(data, 'ward')
    return Z
```

```

from matplotlib import pyplot as plt

Z = calculate_clusters(distance_df)

labels = list(distance_df.columns)
# Extract the membership to a cluster, either specify the n_clusters
# or the cut height
# (similar to sklearn labels)
print(cut_tree(Z, n_clusters=4))
print(Z)

# Visualize the clustering as a dendogram
fig = plt.figure(figsize=(25, 10))
dn = dendrogram(Z, orientation='right', labels=labels)
plt.show()
```

```
def extract_cell_name_from_clusters(Z, n_clusters=5):
    clusters = cut_tree(Z, n_clusters=n_clusters)
    clusters.reshape(clusters.shape[0])

    grouped = {}
    for i, cluster in enumerate(clusters):
        if str(cluster[0]) not in grouped:
            grouped[str(cluster[0])] = []

        grouped[str(cluster[0])].append(labels[i]) 

    grouped_inv = {}
    for (k, v) in grouped.items():
        for cell_name in v:
            grouped_inv[cell_name] = k

    df = pd.DataFrame(data=map(lambda cell_cluster: [cell_cluster[0], cell_cluster[1]], grouped_inv.items()))
    df.columns = ['cell_name', 'cluster']
    return df
```

```
clusters = extract_cell_name_from_clusters(Z)
```
